```python
from torchvision import models
model = models.resnet152(weights="ResNet152_Weights.DEFAULT")
print(model)
```

```python
# count number of convolutional layers
import torch.nn as nn
conv_layers = [layer for layer in model.modules() if isinstance(layer, nn.Conv2d)]
print(len(conv_layers))

# count number of convultional layers in layer3 and layer4
conv_layers = [layer for layer in model.layer3.modules() if isinstance(layer, nn.Conv2d)]
print(len(conv_layers))
conv_layers = [layer for layer in model.layer4.modules() if isinstance(layer, nn.Conv2d)]
print(len(conv_layers))
```
